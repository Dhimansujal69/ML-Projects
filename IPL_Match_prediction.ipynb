```python
df = df.drop('result_margin',axis=1)
```

```python
df = df.drop('method',axis=1)
```

```python
df['target_runs'] = df['target_runs'].fillna(df['target_runs'].mode()[0])
```

```python
df['target_overs'] = df['target_overs'].fillna(df['target_overs'].mode()[0])
```

```python
df['city'] = df['city'].fillna('Unknown')
```

```python
df = df.dropna(subset=['winner'])
```

```python
df.isnull().sum()
```

```python
df = df.drop(['player_of_match','result','target_runs','target_overs','umpire1','umpire2','date'],axis=1)
```

```python
df.info()
```

```python
df.drop('id',axis=1,inplace=True)
```

```python
team_name_map = {
    "Delhi Daredevils": "Delhi Capitals",
    "Deccan Chargers": "Sunrisers Hyderabad",   # optional (or keep separate)
    "Kings XI Punjab": "Punjab Kings",
    "Rising Pune Supergiant": "Rising Pune Supergiants",
    "Royal Challengers Bangalore": "Royal Challengers Bengaluru"
}

# Apply mapping to all team-related columns
team_columns = ["team1", "team2", "toss_winner", "winner"]

for col in team_columns:
    df[col] = df[col].replace(team_name_map)
```

```python
# Total matches played by each team (team1 + team2 count)
team_matches = pd.concat([df["team1"], df["team2"]]).value_counts()

# Total wins by each team
team_wins = df["winner"].value_counts()

# Win rate = wins / matches
team_win_rate = (team_wins / team_matches).fillna(0)

# Add features to dataset
df["team1_win_rate"] = df["team1"].map(team_win_rate)
df["team2_win_rate"] = df["team2"].map(team_win_rate)
```

```python
df["team1_toss_win"] = (df["toss_winner"] == df["team1"]).astype(int)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
columns = ['super_over']
for col in columns:
    df[col] = le.fit_transform(df[col])
```

```python
df['super_over'].value_counts()
```

```python
df.info()
```

```python
df = pd.get_dummies(df,columns=['season','city','match_type','venue','team1','team2','toss_winner','toss_decision'],drop_first=True)
```

```python
print(df.columns)
```

```python
x = df.drop('winner',axis=1)
y = df['winner']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
lr = LogisticRegression()
rf = RandomForestClassifier(random_state=42)
xgb = XGBClassifier()
```

```python
lr.fit(x_train,y_train)
y_pred_lr = lr.predict(x_test)
```

```python
lr.score(x_test,y_pred_lr)
```

```python
print("Accuracy Score of Logistic Regression",accuracy_score(y_test,y_pred_lr))
```

```python
print("Confusiion Matrix of Logistic Regression",confusion_matrix(y_test,y_pred_lr))
```

```python
print("Classification Report of Logistic Regression",classification_report(y_test,y_pred_lr))
```

```python
rf.fit(x_train,y_train)
y_pred_rf = rf.predict(x_test)
```

```python
rf.score(x_test,y_pred_rf)
```

```python
print("Accuracy Score of Random Forest",accuracy_score(y_test,y_pred_rf))
```

```python
print("Confusiion Matrix of Random Forest",confusion_matrix(y_test,y_pred_rf))
```

```python
print("Classification Report of Random Forest",classification_report(y_test,y_pred_rf))
```

```python
from sklearn.model_selection import RandomizedSearchCV
# Hyperparameter search space
param_dist = {
    "n_estimators": [100, 200, 300, 400, 500],
    "max_depth": [None, 10, 20, 30, 40, 50],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "max_features": ["sqrt", "log2", None],
    "bootstrap": [True, False]
}

# Randomized Search
random_search = RandomizedSearchCV(
    estimator=rf,
    param_distributions=param_dist,
    n_iter=30,          # number of random combinations
    cv=5,               # cross validation
    verbose=2,
    random_state=42,
    n_jobs=-1
)

# Fit on training data
random_search.fit(x_train, y_train)

# Best parameters
print("Best Parameters:", random_search.best_params_)

# Best model
best_rf = random_search.best_estimator_

# Predictions
y_pred = best_rf.predict(x_test)

# Accuracy
print("Tuned Random Forest Accuracy:", accuracy_score(y_test, y_pred))

# Report
print(classification_report(y_test, y_pred))
```
