```python
df.drop(['EmployeeCount','EmployeeNumber','Over18','StandardHours'],axis=1,inplace=True)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

binary_columns = ['Gender','OverTime']

for col in binary_columns:
    df[col] = le.fit_transform(df[col])
```

```python
df = pd.get_dummies(df,columns=['MaritalStatus','JobRole','EducationField','Department','BusinessTravel'],drop_first=True)
```

```python
df.value_counts('Attrition')
```

```python
df['Attrition'] = df['Attrition'].replace({'Yes':1,'No':0})
```

```python
df.value_counts('Attrition')
```

```python
df.info()
```

```python
bool_cols = df.select_dtypes(include='bool').columns

df[bool_cols] = df[bool_cols].astype(int)
```

```python
df.info()
```

```python
sns.countplot(x='Attrition' ,data=df)
plt.title("Attrition countplot")
plt.show()
```

```python
sns.countplot(x='OverTime', hue='Attrition', data=df)
plt.title("OverTime vs Attrition")
plt.show()
```

```python
sns.histplot(data=df, x='MonthlyIncome', hue='Attrition', bins=30)
plt.title("MonthlyIncome vs Attrition")
plt.show()
```

```python
df.corr()
```

```python
df.drop('JobLevel',axis=1,inplace=True)
```

```python
df.columns
```

```python
x = df.drop('Attrition',axis=1)
y = df['Attrition']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)

x_train,y_train = smote.fit_resample(x_train,y_train)
```

```python
y_train.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

LR  = LogisticRegression(class_weight='balanced')
LR.fit(x_train_scaled,y_train)
```

```python
y_pred_lr = LR.predict(x_test_scaled)
y_pred_lr
```

```python
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier()
RF.fit(x_train_scaled,y_train)
```

```python
y_pred_rf = RF.predict(x_test_scaled)
y_pred_rf
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier()
DT.fit(x_train_scaled,y_train)
```

```python
y_pred_dt = DT.predict(x_test_scaled)
y_pred_dt
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

print("Accuracy of Logistic Regression: ", accuracy_score(y_test,y_pred_lr))
print("Accuracy of Random Forest: ", accuracy_score(y_test,y_pred_rf))
print("Accuracy of Decision Tree: ", accuracy_score(y_test,y_pred_dt))
```

```python
cm_lr = confusion_matrix(y_test,y_pred_lr)
print("Confusin Matrix of Logistic Regression: \n",cm_lr)
```

```python
sns.heatmap(cm_lr,annot=True,fmt='d')
plt.show()
```

```python
cm_rf = confusion_matrix(y_test,y_pred_rf)
print("Confusin Matrix of Random Forest: \n",cm_rf)
```

```python
sns.heatmap(cm_rf,annot=True,fmt='d',color="Red")
plt.show()
```

```python
cm_dt = confusion_matrix(y_test,y_pred_dt)
print("Confusin Matrix of Decison Tree: \n",cm_dt)
```

```python
sns.heatmap(cm_dt,annot=True,fmt='d',color="Blue")
plt.show()
```

```python
cf_lr = classification_report(y_test,y_pred_lr)
print("Classification Report of Logistic Regression: \n",cf_lr)
```

```python
cf_rf = classification_report(y_test,y_pred_rf)
print("Classification Report of Random Forest: \n",cf_rf)
```

```python
cf_dt = classification_report(y_test,y_pred_dt)
print("Classification Report of Decision Tree: \n",cf_dt)
```
