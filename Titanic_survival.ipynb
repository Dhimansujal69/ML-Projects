```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
df.head(10)
```

```python
df.info()
```

```python
df.shape
```

```python
df.columns
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df = df.drop('Cabin',axis='columns')
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df['Age'] = df["Age"].fillna(df['Age'].mean())
```

```python
df.isnull().sum()
```

```python
df = df.drop(['Name','Ticket','PassengerId'],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in df.columns:
    if df[col].dtype=='object':
        df[col] = le.fit_transform(df[col])

```

```python
df.head(10)
```

```python
y = df['Survived']
```

```python
x = df.drop('Survived',axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test, y_train, y_test = train_test_split(x,y, test_size=0.2 , random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
y_test
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import accuracy_score, r2_score, mean_absolute_error
```

```python
print("MAE: ",mean_absolute_error(y_pred,y_test))
print("Accuracy Score: ",accuracy_score(y_pred,y_test))
print("R2 Score: ",r2_score(y_pred,y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier()

DT.fit(x_train, y_train)
```

```python
y_pred_1 = DT.predict(x_test)
```

```python
DT.score(x_test,y_pred_1)

print("MAE: ",mean_absolute_error(y_pred_1,y_test))
print("Accuracy Score: ",accuracy_score(y_pred_1,y_test))
print("R2 Score: ",r2_score(y_pred_1,y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier()

RF.fit(x_train,y_train)
```

```python
y_pred_2 = RF.predict(x_test)
RF.score(x_test,y_test)

print("MAE: ",mean_absolute_error(y_pred_2,y_test))
print("Accuracy Score: ",accuracy_score(y_pred_2,y_test))
print("R2 Score: ",r2_score(y_pred_2,y_test))
```

```python
feature_importance = pd.Series(RF.feature_importances_,index=x.columns)
feature_importance.sort_values(ascending=False)
```

```python
feature_importance.sort_values().plot(kind='barh')
plt.title("Feature Importance: ")
plt.show()
```

```python
from sklearn.metrics import confusion_matrix

cm=confusion_matrix(y_test, y_pred_2)

print(cm)
```

```python
sns.heatmap(cm , annot=True, fmt='d')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
#Hyperparameter Tunning

from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators':[50,100,200],
    'max_depth':[None,5,10],
    'min_samples_split':[2,5,10]}
```

```python
grid = GridSearchCV(RF, param_grid, cv=5)
grid.fit(x_train, y_train)
```

```python
print(grid.best_params_)
```

```python
best_model = grid.best_estimator_

best_model.score(x_test, y_test)
```
