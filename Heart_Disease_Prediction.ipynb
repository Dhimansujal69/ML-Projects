```python
x = df.drop('target',axis=1)
y = df['target']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)

x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression(class_weight='balanced')

LR.fit(x_train_scaled,y_train)
```

```python
y_pred_lr = LR.predict(x_test_scaled)
y_pred_lr
```

```python
y_test
```

```python
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier()

RF.fit(x_train_scaled,y_train)
```

```python
y_pred_rf = RF.predict(x_test_scaled)
y_pred_rf
```

```python
from sklearn.neighbors import KNeighborsClassifier

KNN = KNeighborsClassifier()

KNN.fit(x_train_scaled,y_train)
```

```python
y_pred_knn = KNN.predict(x_test_scaled)
y_pred_knn
```

```python
from xgboost import XGBClassifier

xg = XGBClassifier()
xg.fit(x_train_scaled,y_train)
```

```python
y_pred_xg = xg.predict(x_test_scaled)
y_pred_xg
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

print("Accuracy of Logistic Regression: ", accuracy_score(y_test,y_pred_lr))
print("Accuracy of Random Forest: ", accuracy_score(y_test,y_pred_rf))
print("Accuracy of KNN: ", accuracy_score(y_test,y_pred_knn))
print("Accuracy of XGBoost: ", accuracy_score(y_test,y_pred_xg))
```

```python
cm_lr = confusion_matrix(y_test,y_pred_lr)
print("Confusin Matrix of Logistic Regression: \n",cm_lr)
```

```python
sns.heatmap(cm_lr,annot=True,fmt='d')
plt.show()
```

```python
cm_rf = confusion_matrix(y_test,y_pred_rf)
print("Confusin Matrix of Random Forest: \n",cm_rf)
```

```python
sns.heatmap(cm_rf,annot=True,fmt='d',color="Red")
plt.show()
```

```python
cm_knn = confusion_matrix(y_test,y_pred_knn)
print("Confusin Matrix of KNN: \n",cm_knn)
```

```python
sns.heatmap(cm_knn,annot=True,fmt='d',color="Red")
plt.show()
```

```python
cm_xg = confusion_matrix(y_test,y_pred_xg)
print("Confusin Matrix of XGBoost: \n",cm_xg)
```

```python
sns.heatmap(cm_xg,annot=True,fmt='d',color="Red")
plt.show()
```

```python
cf_lr = classification_report(y_test,y_pred_lr)
print("Classification Report of Logistic Regression: \n",cf_lr)
```

```python
cf_rf = classification_report(y_test,y_pred_rf)
print("Classification Report of Random Forest: \n",cf_rf)
```
